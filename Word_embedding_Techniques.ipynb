```python
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
### Vocabulary size
voc_size=500
```

#### One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

All the sentences are of different size , if i want to train this in neural network then all the sentences size should be same that is the reason we do post and pre padding.

Here max_size of sentence is 5

### Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## pre padding  -- for all the sentences it will make sure size would be 8 by adding 0 to last or at first
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

Now we will convert each and every value with vectors

```python
## 10 feature dimesnions -- here we are representing each value with 10 dimension
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))  ##vocabulary_size, feature dimension, length of sentence
model.compile('adam','mse')
```

```python
model.summary()
```

```python
##'the glass of milk',
embedded_docs[0]
```

```python
model.predict(embedded_docs[0])   ##here each value is represented by 10 dimension.
```

```python
print(model.predict(embedded_docs))
```

```python
embedded_docs[0]
```

```python
print(model.predict(embedded_docs)[0])
```
